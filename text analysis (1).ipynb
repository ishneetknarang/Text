```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer as SIA
sentences =["great movie that was , acting was amazing"]
sid = SIA()
for sentence in sentences:
    ss = sid.polarity_scores(sentence)
```

```python
def polarity_score(text1):
    sentences = [text1]
    sid = SIA()
    for sentence in sentences:
        ss = sid.polarity_scores(sentence)
    val = ss['pos']-ss['neg']
    return val 
```

```python
data['score'] = text['Q.16 What did you most appreciate / enjoy/think was the best about the course? Any suggestions for improvements. '].apply(polarity_score)
```
